```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
data = pd.read_csv('/Users/aoamacsplace/Documents/mosaic/Aave/Datasets/export-simple-userdata-aave.csv')

df = data.copy()
```

```python
# Total NFTs owned
total_nfts_owned = df['Total NFT`s'].sum()
print('Total NFTs owned: ', total_nfts_owned)

# Distribution of NFT ownership
fig = px.histogram(
    df,
    x='Total NFT`s',
    title='Distribution of NFT Ownership'
)
fig.show()

# NFT ownership vs proposal interaction
fig = px.scatter(
    df,
    x='Total NFT`s',
    y='Proposal Interaction',
    trendline='ols',
    title='NFT Ownership vs Proposal Interaction'
)
fig.show()

# Rewards earned through NFTs
fig = px.box(
    df,
    x='Total Rewards NFT`s',
    title='Distribution of Rewards Earned through NFTs'
)
fig.show()

# Top NFT owners
top_nft_owners = df.nlargest(5, 'Total NFT`s')
fig = go.Figure(data=[go.Table(
    header=dict(values=['Address', 'Total NFT`s', 'Total Votes']),
    cells=dict(values=[top_nft_owners['Address'], top_nft_owners['Total NFT`s'], top_nft_owners['Total Votes']])
)])
fig.update_layout(title='Top 5 NFT Owners')
fig.show()

```
