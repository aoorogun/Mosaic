```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
data = pd.read_csv('/Users/aoamacsplace/Documents/mosaic/Aave/Datasets/export-simple-userdata-aave.csv')

df = data.copy()

df.head()
```

```python
# Average follow count
avg_follow_count = df['Follow Count'].mean()
print('average follow count is:', avg_follow_count)

# Distribution of follow count
fig = px.histogram(
    df,
    x='Follow Count',
    title='Distribution of Follow Count'
)
fig.show()


# Most followed users
top_followed_users = df.nlargest(5, 'Follow Count')
fig = go.Figure(data=[go.Table(
    header=dict(values=['Address', 'Follow Count', 'Total Votes']),
    cells=dict(values=[top_followed_users['Address'], top_followed_users['Follow Count'], top_followed_users['Total Votes']])
)])
fig.update_layout(title='Top 5 Most Followed Users')
fig.show()

# Follow count vs voting behavior
fig = px.box(
    df,
    x='Follow Count',
    y='Proposal Interaction',
    title='Follow Count vs Voting Behavior'
)
fig.show()

# User engagement in specific spaces
fig = px.histogram(
    df,
    x='Follows Space',
    title='User Engagement in Specific Spaces'
)
fig.show()


```
