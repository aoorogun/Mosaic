```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
data = pd.read_csv('/Users/aoamacsplace/Documents/mosaic/Aave/Datasets/export-simple-userdata-aave.csv')

df = data.copy()
```

```python
# Most active users based on proposal interaction
top_active_users = df.nlargest(5, 'Proposal Interaction')
fig = go.Figure(data=[go.Table(
    header=dict(values=['Name', 'Proposal Interaction']),
    cells=dict(values=[top_active_users['Name'], top_active_users['Proposal Interaction']])
)])
fig.update_layout(title='Top 5 Most Active Users')
fig.show()

# Activity comparison: Delegators vs non-delegators
fig = px.histogram(
    df,
    x='Is delegator',
    y='Proposal Interaction',
    color_discrete_sequence=['#636EFA'],
    title='Activity Comparison: Delegators vs Non-Delegators'
)
fig.show()

# User engagement with Ens names
fig = px.histogram(
    df,
    x='Ens name',
    title='User Engagement with Ens Names'
)
fig.show()

# User activity based on avatars
fig = px.histogram(
    df,
    x='Avatar',
    title='User Activity Based on Avatars'
)
fig.show()
```
