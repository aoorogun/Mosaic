```python
import csv
import random
from datetime import datetime, timedelta

# Generate unique values for columns
def generate_unique_values(existing_values, length):
    new_value = ''
    while new_value in existing_values:
        new_value = ''.join(random.choices('0123456789abcdef', k=length))
    existing_values.add(new_value)
    return new_value

# Generate simulated data
def generate_simulated_data(rows):
    wallet_addresses = set()
    ens_names = set()
    ens_telegrams = set()
    ens_twitters = set()
    ens_githubs = set()
    ens_emails = set()

    simulated_data = []
    for _ in range(rows):
        wallet_address = generate_unique_values(wallet_addresses, 40)
        ens_name = generate_unique_values(ens_names, 8)
        ens_telegram = generate_unique_values(ens_telegrams, 8)
        ens_twitter = generate_unique_values(ens_twitters, 8)
        ens_github = generate_unique_values(ens_githubs, 8)
        ens_email = generate_unique_values(ens_emails, 10)

        date_created = datetime.now() - timedelta(days=random.randint(1, 365))
        op_token = random.randint(1, 1000)
        op_tokens_voting = random.randint(1, 500)
        ens_description = 'Simulated description for {}'.format(ens_name)
        ens_url = 'https://example.com/{}'.format(ens_name)
        follow_count = random.randint(1, 1000)
        follow_count_space = random.randint(1, 500)
        follow_count_non_space = random.randint(1, 500)
        total_votes = random.randint(1, 2000)
        total_proposal_interaction = random.randint(1, 100)
        total_proposal_interaction_in_space = random.randint(1, 50)
        total_proposal_interaction_non_space = random.randint(1, 50)
        total_nft = random.randint(1, 50)
        rewards_nft = random.randint(1, 10)

        row = [wallet_address, date_created, op_token, op_tokens_voting, ens_name, ens_description, ens_url,
               ens_telegram, ens_twitter, ens_github, ens_email, follow_count, follow_count_space,
               follow_count_non_space, total_votes, total_proposal_interaction,
               total_proposal_interaction_in_space, total_proposal_interaction_non_space, total_nft, rewards_nft]

        simulated_data.append(row)

    return simulated_data

# Save data to CSV
def save_to_csv(data, filename):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["walletaddress", "date_created", "op_token", "op_tokens_voting", "ensName",
                         "ensDescription", "ensUrl", "ensTelegram", "ensTwitter", "ensGithub",
                         "ensEmail", "followCount", "followCountSpace", "followCountNonSpace",
                         "totalVotes", "totalProposalInteraction", "totalProposalInteractionInSpace",
                         "totalProposalInteractionNonSpace", "totalNft", "rewardsNft"])
        writer.writerows(data)

# Generate simulated data
simulated_data = generate_simulated_data(20000)

# Save data to CSV file
save_to_csv(simulated_data, "./simulated_data.csv")
```
